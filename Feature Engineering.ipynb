### Converting the Hour field in numerical value

```python
data['hour'] = data.hour.str.slice(1,3).astype(int)
```

## HANDLING NUMERICAL DATA

```python
num = data.copy().drop(columns = ['user', 'screen_list', 'first_open', 'enrolled_date', 'enrolled'])
num.head()
```

### HISTOGRAMS

```python
plt.suptitle('Numerical Data', fontsize = 20)
for i in range(1,num.shape[1]-1):
    plt.subplot(3,3,i)
    f = plt.gca()
    f.set_title(num.columns.values[i-1])
    val = np.size(num.iloc[:,i-1].unique())
    plt.hist(num.iloc[:,i-1], bins = val, color = '#3F5D7D')
    
```

### CREATING CORRELATION PLOTS

```python
num.corrwith(data.enrolled).plot.bar(figsize = (20,10),
                                    title = 'Correlation with the Dependent Variable',
                                    fontsize = 15, rot =45,
                                     grid = True
                                    )
```

### BUILDING THE CORRELATION MATRIX

```python
sn.set(style = 'white', font_scale = 2)
corr = num.corr()
mask = np.zeros_like(corr, dtype = np.bool)
mask[np.triu_indices_from(mask)] = True

#Setup Matplotlib Figure
f,axis = plt.subplots(figsize = (18,15))
f.suptitle('Correlation Matrix', fontsize = 40)

# Creating the colormap
cmap = sn.diverging_palette(220,10,as_cmap = True)

# Create heatmap with the mask and aspect ratio
sn.heatmap(corr, mask = mask, cmap = cmap, vmax = 3, center = 0,
          square = True, linewidths=.5, cbar_kws = {'shrink':.5} )
```

## FEATURE ENGINEERING

### CONVERTING THE DATE OBJECT TO DATE TYPE

```python
import dateutil,parser 
data['first_open'] = [dateutil.parser.parse(row_data) for row_data in data['first_open']]
data['enrolled_date'] = [dateutil.parser.parse(row_data) if isinstance(row_data, str) else row_data for row_data in data['enrolled_date']]
```

```python
data.dtypes
```

### CREATING A DIFFERENCE VARIABLE TO FIND THE DIFF BETWEEN OPEN AND ENROLLMENT

```python
data['difference'] = (data.enrolled_date - data.first_open).astype('timedelta64[h]')
```

```python
plt.hist(data['difference'].dropna(), color = '#3FD37D')
plt.title('Distribution of the Time Since Installed')
```

```python
plt.hist(data['difference'].dropna(), color = '#3FD37D', range = [0,100])
plt.title('Distribution of the Time Since Installed')
plt.show()
```

#### REMOVING THE THOSE USERS THAT ENROLLED AFTER 2 DAYS IE; 48 HOURS

```python
data.loc[data.difference > 48, 'enrolled'] = 0
```

```python
data = data.drop(columns = ['enrolled_date', 'difference', 'first_open'])
```

## WORKING ON THE SCREENS ATTRIBUTE

```python
top_screens = pd.read_csv('top_screens.csv').top_screens.values
```

```python
data['screen_list'] = data.screen_list.astype(str) + ','
```

```python
for sc in top_screens:
    data[sc] = data.screen_list.str.contains(sc).astype(int)
    data['screen_list'] = data.screen_list.str.replace(sc+',','')
```

```python
data['other'] = data.screen_list.str.count(',')
data = data.drop(columns = ['screen_list'])
```

```python
# Funnels
savings_screens = ["Saving1",
                    "Saving2",
                    "Saving2Amount",
                    "Saving4",
                    "Saving5",
                    "Saving6",
                    "Saving7",
                    "Saving8",
                    "Saving9",
                    "Saving10"]
data["SavingCount"] = data[savings_screens].sum(axis=1)
data = data.drop(columns=savings_screens)

cm_screens = ["Credit1",
               "Credit2",
               "Credit3",
               "Credit3Container",
               "Credit3Dashboard"]
data["CMCount"] = data[cm_screens].sum(axis=1)
data = data.drop(columns=cm_screens)

cc_screens = ["CC1",
                "CC1Category",
                "CC3"]
data["CCCount"] = data[cc_screens].sum(axis=1)
data = data.drop(columns=cc_screens)

loan_screens = ["Loan",
               "Loan2",
               "Loan3",
               "Loan4"]
data["LoansCount"] = data[loan_screens].sum(axis=1)
data = data.drop(columns=loan_screens)
```
